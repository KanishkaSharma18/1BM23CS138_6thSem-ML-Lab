```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")

# Display first few rows
print(df.head())

# Separate features and target
X = df[['year']]
y = df['per capita income (US$)']

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict income for year 2020
predicted_income = model.predict([[2020]])

print("Predicted per capita income in 2020:", predicted_income[0])

# Plot regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Canada Per Capita Income Prediction")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("salary.csv")

# Check dataset
print(df.head())
print(df.isnull().sum())   # Check missing values

# Remove rows with missing values
df = df.dropna()

# Features and target
X = df[['YearsExperience']]
y = df['Salary']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict salary for 12 years experience
predicted_salary = model.predict([[12]])

print("Predicted salary for 12 years experience:", predicted_salary[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from word2number import w2n

# Load dataset
df = pd.read_csv("hiring.csv")

print(df)

# Convert experience words to numbers
df['experience'] = df['experience'].fillna("zero")
df['experience'] = df['experience'].apply(w2n.word_to_num)

# Fill missing test scores with median
import math
median_test_score = math.floor(df['test_score(out of 10)'].median())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_test_score)

print("\nCleaned Dataset")
print(df)

# Features and target
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions
pred1 = model.predict([[2, 9, 6]])
pred2 = model.predict([[12, 10, 10]])

print("\nPredicted salary for candidate 1:", pred1[0])
print("Predicted salary for candidate 2:", pred2[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("1000_Companies.csv")

print(df.head())

# One-hot encoding for State column
df = pd.get_dummies(df, columns=['State'], drop_first=True)

print("\nAfter Encoding")
print(df.head())

# Features and target
X = df.drop('Profit', axis=1)
y = df['Profit']

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction input
# Florida = 1 in Florida column, others 0
prediction = model.predict([[91694.48, 515841.3, 11931.24, 1, 0]])

print("\nPredicted Profit:", prediction[0])
```
